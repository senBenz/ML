```python
def minmax_scale(x):
  x_norm=[]
  for i in x:
   xi=(i-min(x))/(max(x)-min(x))
   x_norm.append(xi)
   if min(x)==max(x):
     xi=0.0
     x_norm.append(xi)
  return x_norm

print(minmax_scale(x))

```

```python
print("min de x_norm",min(minmax_scale(x)))
print("max de x_norm",max(minmax_scale(x)))
```

```python
def Z_score_pop(x):
  if not x:
    return []

  u = sum(x) / len(x)

  squared_diffs_sum = sum([(val - u)**2 for val in x])
  variance = squared_diffs_sum / len(x)

  if variance == 0:
    return [0.0] * len(x)

  std_dev = variance**0.5
  zscores = []
  for I in x:
    zscore = (I - u) / std_dev
    zscores.append(zscore)

  return zscores

print(Z_score_pop(x))
```
