```python

x = [5.1, 4.9, 6.2, 5.9, 6.7, 5.0, 6.4, 7.0, 5.4, 6.9]
def bubble_sort(x):
  i=0
  for i in range(len(x)):
    for j in range(len(x)-i-1):
      if x[j]>x[j+1]:
        x[j],x[j+1]=x[j+1],x[j]
  return x
print(bubble_sort(x))
sorted_list=bubble_sort(x)
```

```python
def somme(x):
  somme=0
  for i in x :
    somme+=i
  return somme

print(somme(x))
```

```python
def moyenne(x):
  X̄=somme(x)/len(x)
  return X̄
print(moyenne(x))
```
