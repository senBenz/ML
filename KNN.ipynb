```python
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd


df=pd.DataFrame({
    'age':[25,30,22,31,26,45],
    'salary':[2500,3000,2344,2323,3494,90000],
    'premium':[0,1,0,0,1,1]
})


X = df[['age', 'salary']]
y = df['premium']

model=KNeighborsClassifier(n_neighbors=1)
model.fit(X,y)


new_customers = pd.DataFrame({
            'age': [22, 30, 22, 40, 30],
    'salary': [2200, 3200, 2500, 4000, 2200]
})

prediction=model.predict(new_customers)
print(prediction)
```
