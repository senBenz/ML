```python
from sklearn.tree import DecisionTreeClassifier
import pandas as pd

dataf=pd.DataFrame({
    'Heures':[1,2,6,7],
    'res':['echec','echec','reussite','reussite'],

})

X=dataf[['Heures']]

y=dataf['res']

model = DecisionTreeClassifier(max_depth=1)
model.fit(X,y)

new_res=pd.DataFrame({
    'Heures':[8,9,3,0]
})

prediction=model.predict(new_res)

#------------------------------------------------------
#noued de l arbre : si hrs > 5 == reussite  sinon echec 
#------------------------------------------------------

for p in prediction:
    if p == 'echec':
        print("hrs moins de 5")
    else:
        print("hrs plus de 5")


print(prediction)
```
